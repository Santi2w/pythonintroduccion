1.Escribe una función que reciba un número entero positivo como parámetro y devuelva una
lista que contenga los 5 primeros múltiplos de dicho número. Por ejemplo, si la función recibe
el número 3, devolverá la lista [3, 6, 9, 12, 15]. Si la función recibe un parámetro incorrecto
(por ejemplo, un múmero menor o igual a cero), mostrará un mensaje de error por pantalla
y devolverá una lista vacía.

```python
def multiplos(numero):
    if numero <=0:
     print("Error: El numero debe ser un entero positivo:")
     return []
    multiplos = [numero * i for i in range(1,6)]
    return multiplos
    
```

```python
parametro = int(input("Ingresar parametro: "))
print(multiplos(parametro))
    
```

3. Escribir una función donde reciba dos numeros eneteros que sera el rango de una serie. Los datos se debe ingresar por pantalla. y debe mostrar todos los números primos de ese rango

```python
def es_primo(num):
    if num < 2:
        return False
        
    for i in range(2, num):
        if num % i == 0:
            return False 
    return True

def mostrar_primos_en_rango(n1, n2):
    print(f"Números primos entre {n1} y {n2}:")
    contador_primos = 0
    
    for numero in range(n1, n2 + 1):
        if es_primo(numero):
            print(f"El numero {numero} es primo")
            contador_primos += 1
```

```python
n1 = int(input("Ingresar inicio de rango: "))
n2 = int(input("Ingresar fin de rango: "))

mostrar_primos_en_rango(n1, n2)
```

4. Escribir un programa en python donde una función reciba una frase y cuente cuantas vocales tiene la frase.

```python
def cv(frase):
    vocales = 'aeiouAEIOU'
    contador = 0
    for letra in frase: 
        if letra in vocales:
            contador += 1
    return contador 
```

```python
texto = input ('Ingresar texto:')
total = cv(texto)
print(f'La frase contiene {total} vocales: ')
```
