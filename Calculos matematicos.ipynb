Crear un programa en Python donde el usuario ingrese dos numeros enteros por teclado verificar cual de los dos numeros es mayor

```python
n1 = int(input("Ingrese el primer número entero: "))
n2 = int(input("Ingrese el segundo número entero: "))

opciones = [n2, n1]
indice = int(n1 > n2)

print("El número mayor es:", opciones[indice])
```

Ingresar por teclado la fecha actual dia,mes,año y calcular,tambien ingresar la fecha de nacimiento de una persona y calcular cuantos años tiene esa persona en dias,mes y año

```python
dia_actual = int(input("Ingrese el dia actual: "))
mes_actual = int(input("Ingrese el mes actual: "))
año_actual = int(input("Ingrese el año actual: "))

dia_n = int(input("Ingrese el dia de nacimiento: "))
mes_n = int(input("Ingrese el mes de nacimiento: "))
año_n = int(input("Ingrese el año de nacimiento: "))

dias_actuales = año_actual * 365 + mes_actual * 30 + dia_actual
dias_nacimiento = año_n * 365 + mes_n * 30 + dia_n


dias_vividos = dias_actuales - dias_nacimiento
años = dias_vividos // 365
meses = (dias_vividos % 365) // 30
dias = (dias_vividos % 365) % 30

print("\nLa persona tiene:")
print(años, "años")
print(meses, "meses")
print(dias, "días")
```
