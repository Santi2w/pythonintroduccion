```python
#desarollar un programa en python donde se resuelva una ecuacion de segundo grado mediante 
#la formula general el usuario debe ingresar por teclado los valores a,b y c y 
#mostrar en pantalla los valores de la ecuacion
import math

a = float(input("Ingrese el valor de a: "))
b = float(input("Ingrese el valor de b: "))
c = float(input("Ingrese el valor de c: "))

# Validar que no sea división para 0
if a == 0:
    print("Error: a no puede ser 0 porque no es una ecuación de segundo grado.")
else:
    discriminante = b**2 - 4*a*c

    if discriminante > 0:
        x1 = (-b + math.sqrt(discriminante)) / (2*a)
        x2 = (-b - math.sqrt(discriminante)) / (2*a)
        print("La ecuación tiene dos soluciones reales:")
        print("x1 =", x1)
        print("x2 =", x2)

    elif discriminante == 0:
        x = -b / (2*a)
        print("La ecuación tiene una sola solución real:")
        print("x =", x)

    else:
        print("La ecuación no tiene soluciones reales")
```
