### 1 Estructuras de control
1.1 Selección (sentencias condicionales)
* Selección de una de varias alternativas en base a alguna condición.
* Indentación para estructurar el código
* Importante el símbolo ‘:

```python
# # Asigno el bvalo uno a la variable x
x = int(input("Ingrese el valor de x"))
#Implemento una condicional para saber si x es mayor a 0
if x > 0:
 print('Valor positivo')
else:
 print('Valor no positivo')
```

```python
# 3.Escribir un programa en Python para una empresa que tiene sala de juegos para todas las edades y quiere calcular de forma automatica el precio que debe calcular a sus clientes por la entrada.
#El progrma debe preguntar al usuario el nombre y la edad del cliente y mostrar el precio de entrada si el cliente es menor de 4 años puede entrar gratis
#si tiene entre 4 y 18 años debe pagar 5 dolares y si es mañor de 18 años 10 dolares 
# Programa para calcular el precio de entrada a la sala de juegos

nombre = input("Ingrese el nombre del cliente: ")
edad = int(input("Ingrese la edad del cliente: "))

if edad < 4:
    print(f"{nombre}, la entrada es GRATIS.")
elif edad <= 18:
    print(f"{nombre}, el precio de la entrada es $5.")
else:
    print(f"{nombre}, el precio de la entrada es $10.")
```

```python
#Para atributar un determinado impuesto se debe ser mayor a 18 años y 
#tener un ingreso igual o superior de 1000 dolares mensule,escribir un programa que pregunte al usuario su edad y sus ingresos mensuales y muestre por pantalla si el usuario debe atributar o no
# Pedimos los datos
edad = int(input("¿Cuál es tu edad? "))
ingresos = float(input("¿Cuáles son tus ingresos mensuales en dólares? "))

# Primero verificamos si es mayor de edad
if edad > 18:
    # Si es mayor de 18, ahora a verificar sus ingresos
    if ingresos >= 1000:
        print("Debes tributar.")
    else:
        print("No debes tributar (ingresos insuficientes).")
else:
    # Si no es mayor de 18, ni siquiera revisamos el sueldo
    print("No debes tributar (eres menor de edad).")
```
