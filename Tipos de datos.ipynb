# Comentarios de varias lineas
* Texto encapsulado en triples comillas ( que puede ser tanto comillas simples como dobles)
* Se suele usar para documentar bloques de codigo mas significativos

```python
def producto(x,y):
    '''
    Esta funcion recibe dos numeros como parametros y devuelve
    el producto de los mismos
    '''
    return x*y
```

```python
# Llamamos a la funcion
producto(5,6)
```
